```python
import pymongo
from pymongo import MongoClient
import pandas as pd
from datetime import timedelta

# Connect to MongoDB
client = MongoClient("mongodb+srv://marcfolchpomares:AstonMartin1@mycluster.e19nlo1.mongodb.net/ELO_gym?retryWrites=true&w=majority")

db = client.ELO_gym

# Access a specific collection within the database (replace 'your_collection' with your actual collection name)
collection = db.data

# Get the data from the collection
cursor = collection.find({})

# Convert the cursor to a DataFrame
data = pd.DataFrame(list(cursor))

# Convert the 'Date' column to datetime if it's not already
data['Date'] = pd.to_datetime(data['Date'])

# Format the datetime column to include only date and time
data['Date'] = data['Date'].dt.strftime('%Y-%m-%d %H:%M:%S')

# Convert DataFrame to a list of dictionaries (each dictionary represents a document)
data = data.to_dict(orient='records')

# Optionally, drop the existing collection
collection.drop()

# Insert the new data into the collection
collection.insert_many(data)

print("Data replaced successfully.")
```
