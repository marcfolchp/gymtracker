```python
def calculator (weight, repetitions):
    return ((weight/5)*16)+(repetitions-(3*(weight/5)))
```

```python
data2['Actual Score'] = data2.apply(lambda x: calculator(x['Weight (kg)'], x['Repetitions']), axis=1)
```

```python
data2["Goal Score"] = data2["Actual Score"] * multiplier
```
