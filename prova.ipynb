```python
import pandas as pd
import seaborn as sns
import random
import warnings
import numpy as np
```

```python
warnings.filterwarnings("ignore")
```

```python
daterange = pd.date_range(start=pd.to_datetime("01/01/2018"), end=pd.to_datetime("31/01/2018"))
weights = np.random.uniform(low=0.5, high=13.3, size=(31,))
repetitions = [random.randint(1,17)/10 for i in range(0,31)]
weights.sort()
```

```python
pechotriceps = {"Day":daterange, "Weight":weights, "Repetitions":repetitions, "Score":weights+repetitions}
pechotriceps = pd.DataFrame(pechotriceps)
```

```python
sns.lineplot(pechotriceps[["Day", "Score"]]);
```
